```python
import matplotlib.pyplot as plt
import numpy as np

# Sample RMSE, predicted blood glucose, and R-squared values for each model
rmse_values = [221.222738700060237, 16.8341651188332, 0.23651888533479157]  # Example RMSE values for Random Forest, XGBoost, LSTM
predicted_values = [162.56, 159.914794921875, 94.52953338623047]  # Example predicted blood glucose values
r_squared_values = [0.9163466580125496, 0.9568641627182651, 0]  # Example R-squared values

# Models
models = ['Random Forest', 'XGBoost', 'LSTM']

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Bar width
bar_width = 0.25

# Set position of bar on X axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plot bars
plt.bar(r1, rmse_values, color='b', width=bar_width, edgecolor='grey', label='RMSE')
plt.bar(r2, predicted_values, color='g', width=bar_width, edgecolor='grey', label='Predicted BG')
plt.bar(r3, r_squared_values, color='r', width=bar_width, edgecolor='grey', label='R-squared')

# Add xticks on the middle of the group bars
plt.xlabel('Models', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(models))], models)

# Create legend & Show graphic
plt.legend()
plt.show()
```
