```python
import matplotlib.pyplot as plt
import numpy as np

# Sample RMSE, predicted blood glucose, and R-squared values for each model
rmse_values = [21.222738700060237, 16.8341651188332, 0.23651888533479157]  # Example RMSE values for Random Forest, XGBoost, LSTM
predicted_values = [162.56, 159.914794921875, 94.52953338623047]  # Example predicted blood glucose values
r_squared_values = [0.9163466580125496, 0.9568641627182651, 0]  # Example R-squared values

# Models
models = ['Random Forest', 'XGBoost', 'LSTM']

# Number of models
num_models = len(models)

# Width of each bar
bar_width = 0.25

# Set position of bar on X axis
r1 = np.arange(num_models)
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot RMSE for each model
axs[0].bar(r1, rmse_values, color='b', width=bar_width, alpha=0.7)
axs[0].set_ylabel('RMSE')
axs[0].set_title('RMSE for Each Model')
axs[0].set_xticks(range(num_models))
axs[0].set_xticklabels(models)

# Plot Predicted BG for each model
axs[1].bar(r1, predicted_values, color='g', width=bar_width, alpha=0.7)
axs[1].set_ylabel('Predicted Blood Glucose')
axs[1].set_title('Predicted Blood Glucose for Each Model')
axs[1].set_xticks(range(num_models))
axs[1].set_xticklabels(models)

# Plot R-squared for each model
axs[2].bar(r1, r_squared_values, color='r', width=bar_width, alpha=0.7)
axs[2].set_ylabel('R-squared')
axs[2].set_title('R-squared for Each Model')
axs[2].set_xticks(range(num_models))
axs[2].set_xticklabels(models)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```
